```python
# DEŇ 5: RAG v Colab – BEZ CHÝB!
!pip install fpdf2 PyPDF2 google-cloud-aiplatform

from fpdf import FPDF
import vertexai
from vertexai.generative_models import GenerativeModel
import os
import PyPDF2

# 1. Vytvoríme PDF (bez špeciálnych znakov)
pdf = FPDF()
pdf.add_page()
pdf.set_font("Arial", size=12)
pdf.cell(200, 10, txt="AI PROMPT ENGINEER - POPIS PRACE", ln=1, align='C')
pdf.ln(10)
pdf.multi_cell(0, 10,
"""AI Prompt Engineer navrhuje a optimalizuje textove vstupy (prompty) pre velke jazykove modely (LLM) ako Gemini.

Hlavne ulohy:
- Vytvarat efektivne prompty pre rozne ulohy
- Testovat a zlepsovat vystupy AI
- Spolupracovat s vyvojarmi na integracii AI
- Pouzivat GCP na nasadenie rieseni

Zrucnosti:
- Prompt engineering
- Python a GCP
- Kreativita a logicke myslenie""")
pdf.output("AI_Prompt_Engineer_Role.pdf")
print("PDF vytvoreny!")

# 2. RAG – AI číta PDF
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "gemini-key.json"
vertexai.init(project="bamboo-theorem-2000", location="us-central1")
model = GenerativeModel("gemini-2.0-flash")

def read_pdf(file_path):
    with open(file_path, "rb") as f:
        reader = PyPDF2.PdfReader(f)
        text = ""
        for page in reader.pages:
            text += page.extract_text() + "\n"
    return text

pdf_text = read_pdf("AI_Prompt_Engineer_Role.pdf")
print("\nPDF nacitany! Prvych 200 znakov:")
print(pdf_text[:200])

prompt = f"""
Tu je text z dokumentu:
\"\"\"{pdf_text}\"\"\"

Otazka: Ake su hlavne ulohy AI Prompt Engineera?
Odpoved (3 body):
"""

response = model.generate_content(prompt)
print("\nGemini odpoveda (z PDF):")
print(response.text)
```
